```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
trade_with_emotions = pd.read_csv('/content/results.csv')
trade_with_emotions.set_index(trade_with_emotions.columns[0],inplace=True)
```

```python
trade_without_emotions = pd.read_csv('/content/results_wtemo.csv')
trade_without_emotions.set_index(trade_without_emotions.columns[0],inplace=True)
```

```python
def fetch_data(start, end, str):
    # Fetch historical data for the S&P 500 index
    sp500 = yf.Ticker(str)
    data = sp500.history(start=start, end=end)
    return data['Close']

def compute_investment_growth(data, initial_investment):
    # Compute the growth of the investment
    initial_price = data.iloc[0]
    normalized_data = (data / initial_price) * initial_investment
    return normalized_data
```

```python
# Parameters
start_date = '2020-01-01'
end_date = '2024-01-01'
initial_investment = 1000000
sp500 = "^GSPC"
nasdaq = "^IXIC"
# Execution
sp500_prices = fetch_data(start_date, end_date,sp500)
nasdaq_prices = fetch_data(start_date, end_date,nasdaq)
sp500_investment_growth = compute_investment_growth(sp500_prices, initial_investment)
nasdaq_investment_growth = compute_investment_growth(nasdaq_prices, initial_investment)
nasdaq_data = pd.DataFrame(nasdaq_investment_growth)
sp500_data = pd.DataFrame(sp500_investment_growth)
nasdaq_data.index = pd.to_datetime(nasdaq_data.index).tz_convert(None).date
nasdaq_data.index.name = 'date'
sp500_data.index = pd.to_datetime(sp500_data.index).tz_convert(None).date
sp500_data.index.name = 'date'
```
